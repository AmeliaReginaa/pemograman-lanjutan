[TUGAS 2]
Buatlah program/algoritma yang membaca sebuah nilai int positif. Misal N, dan menjumlahkan (serta menampilkan) semua bilangan kelipatan 5 antara 1 s.d. N

```python
N = int(input("Masukkan nilai N: "))

total = 0
kelipatan_5 = []

# Menghitung jumlah dan mengumpulkan bilangan kelipatan 5
for i in range(1, N + 1):
    if i % 5 == 0:
        total += i
        kelipatan_5.append(i)

# Tampilkan hasil
if kelipatan_5:
    print("Output:", total)
    print("Keterangan: Bilangan kelipatan 5 antara 1 s.d.", N, "adalah", kelipatan_5)
    print("Jumlah:", " + ".join(map(str, kelipatan_5)), "=", total)
else:
    print("Output: 0")
    print("Keterangan: Tidak ada bilangan kelipatan 5 antara 1 s.d.", N)
```

[TUGAS 3]
Buatlah 3 versi program yang menerima masukan sebuah integer positif, Misal N (asumsi N>0), dan menuliskan lirik lagu Anak ayam di atas dengan menggunakan perulangan for dan while

```python
# Menggunakan perulangan for

N = int(input("Masukan Jumlah anak ayam :"))

print(f"anak ayam turunlah {N}")
for i in range(N, 0, -1):
    print(f"Mati satu tinggal {i-1}")
print("Mati satu tinggal induknya")
```

```python
# Menggunkan Perulangan while
N = int(input("Masukan Jumlah anak ayam :"))

print(f"anak ayam turunlah {N}")
i = N
while i > 1:
  print(f"Mati satu tinggal {i-1}")
  i -= 1
print("Mati satu tinggal induknya")
```
